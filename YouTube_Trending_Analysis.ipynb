```python
import json

# Load the category JSON file
with open('/Users/keerthidev/Desktop/YouTube Trending Video Analysis/IN_category_id.json') as f:
    data = json.load(f)
```

```python
# Create a dictionary to map ID to category name
category_dict = {}

for item in data['items']:
    cat_id = int(item['id'])  # Category ID (as integer)
    cat_name = item['snippet']['title']  # Category name
    category_dict[cat_id] = cat_name

# Let's print the dictionary to see
print(category_dict)
```

```python
df['category_name'] = df['category_id'].map(category_dict)

# View new column
df[['category_id', 'category_name']].head()
```

```python
print(df.columns)
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce')
```

```python
df['is_all_caps'] = df['title'].apply(lambda x: x.isupper() if isinstance(x, str) else False)

all_caps_count = df['is_all_caps'].sum()
total_videos = df.shape[0]

print(f"{all_caps_count} out of {total_videos} trending videos used ALL CAPS in the title.")
```

```python
df['title_length'] = df['title'].apply(lambda x: len(x) if isinstance(x, str) else 0)

long_titles = df.sort_values(by='title_length', ascending=False).head(5)
print("Top 5 longest video titles:")
print(long_titles[['title', 'title_length']])
```

```python
df['repeated_word'] = df['title'].apply(lambda x: any(word * 2 in x.lower() for word in x.lower().split()) if isinstance(x, str) else False)

print(f"Number of videos with word repetition in the title: {df['repeated_word'].sum()}")
```

```python
import re
emoji_pattern = re.compile("["
         u"\U0001F600-\U0001F64F"  # emoticons
         u"\U0001F300-\U0001F5FF"  # symbols & pictographs
         u"\U0001F680-\U0001F6FF"  # transport & map symbols
         u"\U0001F1E0-\U0001F1FF"  # flags
         "]+", flags=re.UNICODE)

df['emojis'] = df['title'].apply(lambda x: emoji_pattern.findall(x) if isinstance(x, str) else [])
df['emoji_count'] = df['emojis'].apply(len)

# Count total emojis
from collections import Counter
all_emojis = sum(df['emojis'], [])
emoji_freq = Counter(all_emojis).most_common(10)

print("Top 10 emojis used in trending video titles:")
print(emoji_freq)
```

```python
clickbait_keywords = ['you won\'t believe', 'gone wrong', 'shocking', 'must watch', 'epic fail']
pattern = '|'.join(clickbait_keywords)

df['clickbait'] = df['title'].str.contains(pattern, case=False, na=False)

print(f"Total clickbait titles: {df['clickbait'].sum()}")
```

```python
actors = ['vijay', 'yash', 'shahrukh', 'ajith', 'surya', 'alluarjun', 'ranveer', 'prabhas']
df['has_star'] = df['title'].str.lower().apply(lambda x: any(star in x for star in actors) if isinstance(x, str) else False)

star_videos = df[df['has_star']]
print("Number of trending videos featuring movie stars:", star_videos.shape[0])
```

```python
# Calculate average views per publish hour
avg_views_by_hour = df.groupby('publish_hour')['views'].mean().reset_index()

# Plot the average views by publish hour
plt.figure(figsize=(12,6))
sns.lineplot(data=avg_views_by_hour, x='publish_hour', y='views', marker='o', color='royalblue')
plt.title('Average Views by Publish Hour')
plt.xlabel('Publish Hour (0-23)')
plt.ylabel('Average Views')
plt.xticks(range(0,24))
plt.show()
```
